```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs


X, y = make_blobs(n_samples=10_000, n_features=2, centers=5, cluster_std=[1, 0.8, 0.5, 0.5, 0.5], random_state=42)
```

```python
y
```

```python
plt.figure(figsize=(3, 2))
plt.scatter(X[:, 0], X[:, 1], c=None, s=1)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$", rotation=0)
plt.gca().set_axisbelow(True)
plt.grid()
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, init="random", n_init=10, random_state=42)
kmeans.fit(X)
```

```python
kmeans.labels_.shape
```

```python
kmeans.cluster_centers_
```

```python
plt.figure(figsize=(3, 2))
plt.scatter(X[:, 0], X[:, 1], c=kmeans.labels_, s=1)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c="r", s=12)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$", rotation=0)
plt.gca().set_axisbelow(True)
plt.grid()
plt.show()
```

```python
kmeans.inertia_
```

```python
kmeans.n_iter_
```

```python
kmeans.labels_
```

```python
from sklearn.metrics import silhouette_score

silhouette_score(X, kmeans.labels_)
```
